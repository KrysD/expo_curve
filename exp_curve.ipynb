```python
import numpy as np
import matplotlib.pyplot as plt

def draw_exponential_curve(N, k):
    x = np.linspace(0, 10, 1000) # generate 1000 points between 0 and 10
    y = N * np.exp(-k * x) # compute y as N * exp(-k*x)
    plt.plot(x, y) # plot the curve
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Exponential Curve')
    plt.show()
```

In this modified function, we use N to define the initial value of y. We also use the constant k to control the rate of decay of y. The larger the value of k, the faster the curve decreases.
For example, to draw an exponential curve that decreases from N=10 to 0 with k=1, you can execute draw_exponential_curve(10, 1).

```python
draw_exponential_curve(10,1)
```

```python
draw_exponential_curve(50,1)
```
